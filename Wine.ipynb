```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="muted")

cols = df.columns[:-2]  

def plot_feature_distribution(df, col):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f"{col} Analizi", fontsize=16, fontweight='bold')

    sns.histplot(df, x=col, ax=axes[0], bins=30, kde=True, color="skyblue")
    axes[0].set_title(f"{col} - Histogram", fontsize=12)
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Frequency")

    sns.boxplot(data=df, x=col, ax=axes[1], color="lightcoral")
    axes[1].set_title(f"{col} - Boxplot", fontsize=12)
    axes[1].set_xlabel(col)
    axes[1].set_ylabel("Value")

    sns.kdeplot(data=df, x=col, hue="quality", fill=True, ax=axes[2])
    axes[2].set_title(f"{col} - KDE by Quality", fontsize=12)
    axes[2].set_xlabel(col)
    axes[2].set_ylabel("Density")

    plt.tight_layout()
    plt.show()

for col in df.columns[:-2]: 
    plot_feature_distribution(df, col)
```

```python
#Uçucu asitlik değeri arttıkça quality değeri düşer
plt.figure(figsize=(12,8))
sns.barplot(data=df,x="quality",y="volatile acidity",legend=False,palette="viridis")
plt.title("Volatile acidity vs quality")
plt.show()
```
